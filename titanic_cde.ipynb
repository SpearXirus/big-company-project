```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.head(2)
```

```python
titanic = titanic.loc[:, ['sex', 'age', 'class', 'embark_town', 'survived', ]]
```

```python
titanic = titanic.dropna(subset=['age'], how='any', axis=0)
```

```python
# 3-c
df = titanic.loc[:, ['sex', 'class', 'embark_town', 'survived']]
df_sex_avg = df.pivot_table(columns='sex', aggfunc='mean')
df_sex_avg
```

```python
df_class_avg = df.pivot_table(index='class', aggfunc='mean')
df_class_avg
```

```python
df_embark_town_avg = df.pivot_table(index='embark_town', aggfunc='mean')
df_embark_town_avg
```

```python
# 3-d
titanic['연령대'] = pd.cut(x=titanic['age'], # 데이터 배열
                      bins=[1, 20, 30, 50, 70, 100], # 경계값
                      labels=['미성년자', '청년', '중년', '장년', '노년'], # bin이름
                      include_lowest=True)
df1 = titanic.pivot_table(index='연령대', aggfunc='mean')
df1
```

```python
# 3-e
titanic['age_group'] = pd.qcut(titanic['age'],3)
titanic.pivot_table(index='age_group', columns='sex', aggfunc='mean')
```
