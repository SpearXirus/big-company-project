```python
setosa25=setosa.quantile(q=0.25, axis=0)
setosa75=setosa.quantile(q=0.75, axis=0)
versicolor25=versicolor.quantile(q=0.25, axis=0)
versicolor75=versicolor.quantile(q=0.75, axis=0)
virginica25=virginica.quantile(q=0.25, axis=0)
virginica75=virginica.quantile(q=0.75, axis=0)
```

```python
#이상치
def dropsetosa(x):
  if setosa25 - 1.5 * (setosa75 - setosa25)> x:
    x.drop()
  elif setosa75 + 1.5 * (setosa75 - setosa25)< x:
    x.drop()
  else:
    return x
```

```python
def dropversicolor(x):
  if versicolor25 - 1.5 * (versicolor75 - versicolor25)> x:
    x.drop()
  elif versicolor75 + 1.5 * (versicolor75 - versicolor25)< x:
    x.drop()
  else:
    return x
```

```python
def dropvirginica(x):
  if virginica25 - 1.5 * (virginica75 - virginica25)> x:
    x.drop()
  elif virginica75 + 1.5 * (virginica75 - virginica25)< x:
    x.drop()
  else:
    return x
```
